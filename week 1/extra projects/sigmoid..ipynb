```python
import numpy as np 
import matplotlib.pyplot as plt

def sigmoidFx(x):
  return 1 /(1+np.exp(-x))

x = np.linspace(-1,10,100) #NumPy function that creates a list of numbers evenly spaced between -10 and 10.
plt.plot(x,sigmoidFx(x))
plt.xlabel("x values")
plt.ylabel("signmoid value for x")
plt.title("Signmoid fxn")
plt.savefig("sigmoid_plt.png")
plt.show()
print("signmoid(0) =" , sigmoidFx(0))
```
