```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('./deeplearning.mplstyle')
```

```python
x_train = np.array([1.0 , 2.0])
y_train = np.array([300.0 ,500.0])
print(f"x_train = {x_train}")
```

```python
plt.scatter(x_train,y_train,marker='x',c='r')
plt.title("House Prices")
plt.ylabel("price (in 1000s of dollars)")
plt.xlabel('Size (1000 sqft)')
plt.show()
```

```python
# ploting y = mx+c , model funciton
w = 100
b = 100
print(f"w: {w}")
print(f"b: {b}")
```

```python
def compute_model_output(x,w,b):
  """Computes the prediction of a linear model
  Args:
    x (ndarray (m,)):Data , m examples
    w,b (scalar)    : model parameters
  Returns
    y (ndarray (m,)) :target values
  """

  m = x.shape[0] #rows
  f_wb = np.zeros(m) #1 d array with n entries
  for i in range(m):
    f_wb[i] = w*x[i] +b  # filling empty array with the values of mx+c
  return f_wb
```

```python
temp_f_wb = compute_model_output(x_train,200,100)
print(f"w={w} b={b}")

plt.plot(x_train,temp_f_wb,c='b',label='Our Prediction')

plt.scatter(x_train , y_train ,marker='x' , c='r' , label='Actual Values')

plt.title("Housing Prices");

plt.ylabel('Price (in 1000s of dollars)');

plt.xlabel('Size (1000 sqft)');
plt.legend();
plt.show()
```
